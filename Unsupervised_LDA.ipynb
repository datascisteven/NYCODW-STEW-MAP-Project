## Importing packages

```python
import os
import pickle
import pandas as pd
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
os.chdir('/Volumes/My Backup Barracuda/Dropbox/Data Science/DRepositories/NYCODW-STEW-MAP-Project')
```

```python
with open('processed_tweets.pickle', 'rb') as read_file:
    df = pickle.load(read_file)

tokens_list_of_lists = list(df.tweet.values)
tokens_list = df.tweet.sum()
tokens_list_of_strings = [",".join(x).replace(",", " ") for x in tokens_list_of_lists]
```

```python
vectorizer = CountVectorizer()
tf = vectorizer.fit_transform(tokens_list)
tf_feature_names = vectorizer.get_feature_names()
```

## n_components = 3

```python
lda_3 = LatentDirichletAllocation(n_components=3, random_state=42)
lda_3.fit(tf)
```

```python
def display_topics(model, feature_names, no_top_words):
    topic_dict = {}
    for topic_idx, topic in enumerate(model.components_):
        topic_dict["Topic %d words" % (topic_idx)]= ['{}'.format(feature_names[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
        topic_dict["Topic %d weights" % (topic_idx)]= ['{:.1f}'.format(topic[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
    return pd.DataFrame(topic_dict)
```

```python
no_top_words = 10
display_topics(lda_3, tf_feature_names, no_top_words)
```

```python
pyLDAvis.sklearn.prepare(lda_3, tf, vectorizer)
```
