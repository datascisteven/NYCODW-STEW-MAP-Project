```python
import pickle
with open('processed_tweets.pickle', 'rb') as read_file:
    df = pickle.load(read_file)

tokens_list_of_lists = list(df.tweet.values)
tokens_list = df.tweet.sum()
tokens_list_of_strings = [",".join(x).replace(",", " ") for x in tokens_list_of_lists]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_df=0.9, 
                             min_df=25,
                             token_pattern='\w+|\$[\d\.]+|\S+',
                             binary=True,
                             stop_words='english')

dtm_tf = vectorizer.fit_transform(tokens_list)
```

```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_3 = LatentDirichletAllocation(n_components=3, random_state=42)
lda_3.fit(dtm_tf)
```

```python
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda_3, dtm_tf, vectorizer)
```
